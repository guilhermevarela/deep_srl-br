 # 1. DEFINE I/O DATA FRAMES

 ## Specifications can be found @
 > http://www.lsi.upc.edu/~srlconll/soft.html#srlconll

```python
import numpy as np 
import pandas as pd 
import re 

MODEL_NAME='multi_bibasic_lstm'
HPARAMS_STR='lr5.00e-04_hs128x64'
MODEL_V= '01'
INPUT_DIR= '../datasets/csvs/'
OUTPUT_DIR='../outputs/'

INPUT_PATH = '{:}{:}'.format(INPUT_DIR, 'zhou_valid.csv')
OUTPUT_PATH= '{:}{:}/{:}/{:}/{:}'.format(OUTPUT_DIR,MODEL_NAME,HPARAMS_STR,MODEL_V, 'Yhat_valid.csv')
TARGET_EVALUATION_PATH= '{:}{:}/{:}/{:}/{:}'.format(OUTPUT_DIR,MODEL_NAME,HPARAMS_STR,MODEL_V, 'conll_evaluate_valid.txt')
TARGET_GOLDENSTD_PATH=  '{:}{:}/{:}/{:}/{:}'.format(OUTPUT_DIR,MODEL_NAME,HPARAMS_STR,MODEL_V, 'conll_golden_valid.txt')



print(INPUT_PATH)
print(OUTPUT_PATH)
```

```python
df_input= pd.read_csv(INPUT_PATH, index_col=0)
df_output=pd.read_csv(OUTPUT_PATH, index_col=0)
print(df_input.shape, df_output.shape)
```

```python
df= df_input.join(df_output, how='left')
print(df.shape)
```

```python
df.head()
```

 # 2. Make Evaluation

```python
df_evaluation= df[['S', 'P', 'FUNC', 'Y_0']]
# re.sub(pattern, repl, string, count=0, flags=0)

#Replace '-' making it easier to concatenate
sub_fn= lambda x: re.sub('-', '', x)
df_evaluation['FUNC'] = df_evaluation['FUNC'].apply(sub_fn)


s0= min(df['S'])
sn= max(df['S'])
for i,si in enumerate(range(s0,sn+1)):
    dfsi=df_evaluation[ df_evaluation['S'] == si ]
    p0= min(dfsi['P'])
    pn= max(dfsi['P'])    
    for j,p in enumerate(range(p0,pn+1)): # concatenate by argument columns
        dfpj=dfsi[dfsi['P']==p].reset_index(drop=True)
        if j==0:
            dfp=dfpj[['FUNC','Y_0']]
            dfp=dfp.rename(columns={'Y_0': 'ARG0'})
        else:
            dfp['FUNC']=dfp['FUNC'].map(str).values + dfpj['FUNC'].map(str).values
            dfp = pd.concat((dfp, dfpj['Y_0']), axis=1)
            dfp=dfp.rename(columns={'Y_0': 'ARG{:d}'.format(p-p0)})            
    if i==0:        
        dfconll_evaluation= dfp
    else:
        dfconll_evaluation= dfconll_evaluation.append(dfp).fillna('')

# dfconll_evaluation.to_csv(TARGET_EVALUATION_PATH, sep= '\t', index=False, header=False)
import code; code.interact(local=dict(globals(), **locals()))		
num_columns=dfconll_evaluation.columns
usecolumns=['FUNC'] + ['ARG{:d}'.format(i) for i in range(num_columns-1)]
dfconll_evaluation=dfconll_evaluation[usecolumns]
dfconll_evaluation.to_csv(TARGET_EVALUATION_PATH, sep= '\t', index=False, header=False)
dfconll_evaluation.head() 
```

 # 3. Make Gold

```python
dfconll_golden= df[['FUNC', 'ARG_0']]
dfconll_golden.to_csv(TARGET_GOLDENSTD_PATH, sep= '\t', index=False, header=False)
```
