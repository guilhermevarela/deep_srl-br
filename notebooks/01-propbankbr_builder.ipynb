```python
# This dataframe has all multiple arguments on columns ARG0 ~ARG6
df = propbankbr_parser()
df.head()
```

 ### 1. 1 Normalization
Convert each record into a machine learning example i.e one argument per row
* Filter        .: each ARG column
* Create        .: tmp row holding the number of the argument ( for further ordering )
* Concatenate   .: everybody afterwards by rows
* Sort          .: reorder by S and TMP


```python
columns_features= df.columns[:12]
df_feature= df.loc[:,columns_features].copy()

dataframes=[]
num_records=0
for i in range(7):
    col= 'ARG{:}'.format(i)
    df_target= df[col].dropna().to_frame()
    df_arg = df_feature.join(df_target, how='right')
    df_arg = df_arg.rename(index=str, columns={col: 'ARG'})
    df_arg['TMP']=i+1
    
    num_records+=df_arg.shape[0]
    print('{:}-\tnew records: {:}\ttotal records: {:}'.format(i, df_arg.shape[0],num_records))
    dataframes.append(df_arg)

#Concatente, sort by sentence & argument, reindex
df_norm = pd.concat(dataframes,axis=0)
df_norm.sort_values(by=['S','TMP'],axis=0 ,kind='mergesort', inplace=True)    
df_norm= df_norm.reset_index(drop=True)
df_norm.index.names=['INDEX']
```

 ### 1.2 FIXING EXISTING FEATURES 
 #### 1.2.1 FIXING PREDICATE
  * __PRED__ must be non empty only when ARG is (V*) 
  * Drop column __TMP__

```python
#Fix predicate 
index= (df_norm['TMP'] == df_norm['P_S'])
df_norm.loc[~index,'PRED']= '-'
df_norm.drop(labels='TMP', axis=1, inplace=True)
# DataFrame.drop(labels=None, axis=0, index=None, columns=None, level=None, inplace=False, errors='raise')
df_norm.head()
```

 #### 1.2.2 FIXING P
  __P__ will be a range like id over the predicates.

```python
P=[]
P_S=[]
S= list(set(df_norm['S'].values))
p=1

for s in S:
    sindex=df_norm['S'] == s
    ind= [int(x) for x in 
        df_norm.loc[sindex,'ID'].values]
    p_s= [int(x) for x in 
        df_norm.loc[sindex,'P_S'].values]
    l =max(ind)
    n_p= len(set(filter(lambda x : x > 0, p_s)))
    # redefining p_s
    for p_s in range(n_p):
        P+= l*[p]
        P_S+= l*[p_s]
        p+=1

df_norm['P']=P        
df_norm['P_S']=P_S        
df_norm.head()                    
```

### 2.1 M_R

m_r or region mark denotes the argument position m_r=1 if it locates in the predicate context region, or m_r= 0 if not.

```python
P= list(set(df_norm['P'].values))
M_R=[]
for p in P:
    ind_p=(df_norm['P'] == p)
    ind_p= (df_norm.loc[ind_p, 'PRED'] != '-').values
    M_R+=list(np.cumsum(ind_p))


print(len(M_R))
M_R[:10]
```

 ### 2.2 CTX_P

```python
CTX_P={}
LAGS=list(filter(lambda x : x != 0,list(range(-3,4))))
P= list(set(df_norm['P'].values))
for lag in LAGS:        
    CTX_P[lag]=[]   

for p in P:    
    ind_p=(df_norm['P'] == p)
    n_p= sum(ind_p)
    
    ispredicate=(df_norm.loc[ind_p, 'PRED'] != '-').values    
    #Should be only one
    idx=min([i for i,x in 
                enumerate(ispredicate) if x])

    LEMMA_P=list((df_norm.loc[ind_p, 'LEMMA']).values)
    prev_lag=0
    for lag in LAGS:        
        if ((idx + lag) >-1) and ((idx + lag) < n_p):
            CTX_P[lag]+=[LEMMA_P[(idx + lag)]]*n_p 
        else:
            CTX_P[lag]+=['-']*n_p 

    if (p % 50 == 0):         
        buffer= 'processed propositions:{:5d}\tcomplete:{:0.2f}%\r'.format(p, 100*float(p)/len(P))        
        sys.stdout.write(buffer)
        sys.stdout.flush()        

buffer= 'processed propositions:{:5d}\tcomplete:{:0.2f}%\r'.format(P[-1], 100*float(P[-1])/len(P))        
print(buffer)
print(len(CTX_P[3]))
print(CTX_P[-3][:25])
```

 ### 2.3 PRED_1
Is a transformation on the column __PRED__ build by repeating the predicate over the sequence.
 
Why:

    'Since a single predicate word can not exactly describe thepredicate  information,  especially  when  the  samewords appear more than one times in a sentence.With the expanded context, the ambiguity can belargely eliminated.' (ZHOU, et XU, 2015)

```python
P= list(set(df_norm['P'].values))
predindex=(df_norm.loc[:, 'PRED'] !=  '-').values
PRED_p= df_norm.loc[predindex, 'PRED'].values
PRED_1=[]

for p in P:
    ind_p=(df_norm['P'] == p)
    p_len = sum(ind_p)        
    PRED_1 += [PRED_p[p-1]] * p_len

    if (p % 50 == 0):         
        buffer= 'processed propositions:{:5d}\tcomplete:{:0.2f}%\r'.format(p, 100*float(p)/len(P))        
        sys.stdout.write(buffer)
        sys.stdout.flush() 

buffer= 'processed propositions:{:5d}\tcomplete:{:0.2f}%\r'.format(P[-1], 100*float(P[-1])/len(P))        
print(buffer)        
print(len(PRED_1))    
print(PRED_1[:35])    
```

 ## 2.4 T
 Performs a transformation over ARG in order to make easier predictions for softmax
 
    __Markups:__
     [A0 He ] [AM-MOD would ] [AM-NEG n't ] [V accept ] [A1 anything of value ] from [A2 those he was writing about ]
 
     __ARG:__
      (\A0*)       (\*AM-MOD)    (\*AM-NEG)       (\*V)              (A1\* * *)   *   (A2* * * * *)
      
     __T:__
         A0           AM-MOD        AM-NEG           V                A1 A1 A1    *    A2 A2 A2 A2 A2

```python
def propbankbr_make_targets(propositions, arguments):
	'''
        Returns a T column with a transformation over arguments
        
        args:
            propositions .:  list<string> propostion index (P column)
            arguments    .:  list<string> arguments (ARG column)
            
        returns:
            targets      .: list<string> targets
	'''
	prev_tag=''
	prev_prop=-1
	targets=[]
	for prop, tag in zip(propositions, arguments):
		if prev_prop == prop: 
			if (tag in ['*']): # Either repeat or nothing
				if (')' in prev_tag): 
					target='*'
				else:	
					target=targets[-1] # repeat last
			else:
				if  (')' in prev_tag): #last tag is closed, refresh					 
					target=re.sub(r'\(|\)|\*|','',tag)					
				else:
					if prev_tag != tag and tag != '*)':
						target=re.sub(r'\(|\)|\*|','',tag)					
					else:
						target=targets[-1]			
		else: 
			if (tag in ['*']):
				target='*'
			else:
				target=re.sub(r'\(|\)|\*|','',tag)

		prev_tag= tag 	
		prev_prop= prop		
		targets.append(target)
	return targets				
```

```python
P= list(df_norm['P'].values)
ARG= list(df_norm['ARG'].values)

T= propbankbr_make_targets(
    P,
    ARG
)

print(len(T))
print(ARG[:33])
print(T[:33])
```
