<h1><center>Weighted interval Scheduling</center></h1>

Aqui mostramos a implementação em O(n lg(n)) do **weighted interval scheduling(WIS)**


## Problema

Um problema WIS contém 3 vetores, tempos inciais $s_i$, tempos finais $f_i$ e valores dos intervalos $v_i$

Queuremos o subconjunto de intervalos sem sobreposição que maximizam o valor

```python
import numpy as np
import utils_view as uv
import weighted_interval_scheduling as wis
```

Criamos um conjunto problema com alguns intervalos

```python
s = np.array([0,2,3,5,7,11,13])
f = np.array([5,4,7,8,12,14,15])
v = np.array([3,3,2,2,5,4,3,4])
    
uv.show_intervals(s,f,v)
```

Executamos o algoritmo de *Weighted Interval Scheduling* a fim de obter o cronograma ótimo de intervalos sem sobreposição.

O método **compute_schedule** retorna os índices dos intervalos escolhidos no agendamento

```python
r = wis.compute_schedule(s,f,v)
print(r)
rfilter = np.array([False]*len(s))
rfilter[r] = True
uv.show_intervals(s, f,v,rfilter)
```

## Exemplo Aleatório

```python
n = 10

s = np.random.uniform(0,n,n)
f = np.random.uniform(s+1,s+n//2,n)
v = np.random.uniform(0,n,n)
```

```python
uv.show_intervals(s,f,v)
```

```python
r = wis.compute_schedule(s,f,v)
```

```python
rfilter = np.array([False]*len(s))
rfilter[r] = True
```

```python
print('schedule score:', np.sum(v[r]))
```

```python
uv.show_intervals(s,f,v,rfilter)
```
